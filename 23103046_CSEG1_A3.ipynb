```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('sales.csv')
```

```python
# Plot histogram for total_profit
plt.figure(figsize=(10, 6))
plt.hist(data['total_profit'], bins=10, color='blue', edgecolor='black')
plt.title('Distribution of Total Profit Across All Months')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
data.boxplot(column=products)
plt.title('Distribution of Monthly Sales (total_units) for Each Product')
plt.xlabel('Product')
plt.ylabel('Total Units Sold')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
for product in products:
    plt.bar(data['month_number'], data[product], label=product)
plt.title('Monthly Sales Figures of All Products')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
bottom = [0] * len(data['month_number'])
for product in products:
    plt.bar(data['month_number'], data[product], bottom=bottom, label=product)
    bottom = [i + j for i, j in zip(bottom, data[product])]
plt.title('Stacked Monthly Sales Figures of All Products')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.legend()
plt.show()
```

```python
total_sales = data[products].sum()
plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=products, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Total Sales by Product')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(total_sales, labels=products, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
plt.title('Doughnut Chart of Total Sales by Product')
plt.show()
```

```python
selected_products = ['facecream', 'facewash', 'toothpaste']
plt.figure(figsize=(14, 8))
bar_width = 0.2
months = data['month_number']

for i, product in enumerate(selected_products):
    plt.bar(months + i * bar_width, data[product], width=bar_width, label=product)

plt.title('Grouped Bar Chart of Monthly Sales for Facecream, Facewash, and Toothpaste')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.xticks(months + bar_width, months)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(data['month_number'], data['total_units'], s=data['total_profit']/100, alpha=0.5)
plt.title('Bubble Chart Representing Monthly Total Units Sold')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.show()
```

```python
import matplotlib.dates as mdates

# Hypothetical campaign data (start and duration)
campaigns = {
    'facecream': (1, 2),
    'facewash': (3, 2),
    'toothpaste': (5, 3),
    'bathingsoap': (2, 4),
    'shampoo': (4, 1),
    'moisturizer': (7, 2)
}

fig, ax = plt.subplots(figsize=(10, 8))

for i, (product, (start, duration)) in enumerate(campaigns.items()):
    ax.broken_barh([(start, duration)], (i-0.4, 0.8), facecolors='tab:blue')

ax.set_yticks(range(len(campaigns)))
ax.set_yticklabels(campaigns.keys())
ax.set_xlabel('Month')
ax.set_title('Gantt Chart of Sales Campaigns')
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.show()
```
